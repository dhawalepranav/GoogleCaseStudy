#### Creating a new column "Trip duration" in dataframe.

```python
df['Trip_duration']=df['ended_at']-df['started_at']
df['Trip_duration']=df['Trip_duration'].apply(lambda x: x.total_seconds()/60)
```

```python
df[df.Trip_duration<0].count()
```

### Removing the data with negative Trip durations.

```python
index=df[df.Trip_duration<0].index
df.drop(index,inplace=True)
df=df.reset_index(drop=True)
```

#### Creating day_of_week column in dataframe which will save a day on which ride was started.

```python
df['day_of_week']=df['started_at'].apply(lambda x:x.weekday())
```

```python
member=df[df.member_casual=='member']
casual=df[df.member_casual=='casual']
member=member.reset_index(drop=True)
casual=casual.reset_index(drop=True)
```
